```python
def perc_change(df):
    returns_df = df.pct_change()
    average_daily_growth = returns_df.mean()
    average_daily_growth = average_daily_growth * 100
    display(average_daily_growth)
    min_change = average_daily_growth.min()
    max_change = average_daily_growth.max()
    median_change = average_daily_growth.median()
    return min_change, max_change, median_change
```

```python
def aggregate_plot(df, sector):
    
    new_df = df.copy()
    new_df = pd.DataFrame(new_df.sum(axis=1))
    new_df.columns = ['Value']
    min_df = float(new_df.min())
    max_df = float(new_df.max())
    ind_min = new_df.loc[new_df['Value'] == min_df].reset_index()['Date'][0]
    ind_max = new_df.loc[new_df['Value'] == max_df].reset_index()['Date'][0]
    
    per_0 = '2016-01-01'
    per_1 = '2017-12-31'
    per_2 = '2019-12-31'
    per_3 = '2021-12-31'
    
    
    new_df.plot()
    
    plt.axvline(per_0, alpha=0.7, color='blue')
    plt.axvline(per_1, alpha=0.7, color='blue')
    plt.axvline(per_2, alpha=0.7, color='blue')
    plt.axvline(per_3, alpha=0.7, color='blue')
    plt.axvline(ind_min, alpha=0.7, color='red')
    plt.axvline(ind_max, alpha=0.7, color='green')
    
    plt.title(f'Aggregate information about {sector}')
    plt.ylabel('Value')
    plt.savefig(f'{sector}.png')
```

```python
aggregate_plot(data_fintech, 'Fintech')
```

```python
aggregate_plot(data_IT, 'IT')
```

```python
aggregate_plot(data_retail, 'Retail')
```

```python
aggregate_plot(data_oil, 'Oil')
```

```python
aggregate_plot(data_estate, 'Real estate')
```

```python
def plot_periods(df, sector):
    df_all = df.copy()
    df1 = df_all.loc[df_all['Date'] < '2016-01-01']
    df2 = df_all.loc[df_all['Date'].between('2016-01-01', '2017-12-31')]
    df3 = df_all.loc[df_all['Date'].between('2018-01-01', '2019-12-31')]
    df4 = df_all.loc[df_all['Date'].between('2020-01-01', '2021-12-31')]
    df5 = df_all.loc[df_all['Date'] > '2022-01-01']
    
    assert len(df1) + len(df2) + len(df3) + len(df4) + len(df5) == len(df_all), 'incorrect'
    
    df1 = df1.set_index('Date')
#     df2 = df2.set_index('Date')
#     df3 = df3.set_index('Date')
#     df4 = df4.set_index('Date')
#     df5 = df5.set_index('Date')
    df1.plot()
    plt.ylabel('Price')
    plt.title(f'Sector: {sector}, Period: from to 2014 to end of 2015')
    plt.show()
```
